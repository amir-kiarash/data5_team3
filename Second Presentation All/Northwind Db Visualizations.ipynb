```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
numberofterritories_df = pd.read_csv('ER1_numberofterritories.csv', names= ['Employee','Title', 'Number of Territories', 'Territories'])
performance_df = pd.read_csv('ER2_performance.csv', names = ['Employee', 'Sales', 'Performance'])
leastsaledcategories_df = pd.read_csv('ER3.csv', names = ['Regions', 'Categories'])
```

```python
numberofterritories_df
```

```python
sns.barplot(data=numberofterritories_df.sort_values(by = 'Number of Territories', ascending = False), x="Employee", y="Number of Territories")
plt.xticks(rotation=45)
```

```python
performance_df
```

```python
sns.barplot(data=performance_df.sort_values(by = 'Sales', ascending = False), x="Employee", y="Sales")
plt.xticks(rotation=45)
```

```python
corr_df = pd.merge(performance_df[['Employee' ,'Sales']], 
         numberofterritories_df[['Employee' ,'Number of Territories']], on = 'Employee')
```

```python
corr_df
```

```python
corr_df
```

```python
# Keyword arguments for styling the plot
kwargs = dict (linestyle='--', color='b', marker ='o', linewidth=1.2, markersize=13)

# Draw the plot
line = sns.scatterplot(x = 'Number of Territories', y = 'Sales', data=corr_df)

for i in range(corr_df.shape[0]):
    plt.text(x=corr_df['Number of Territories'][i]+0.3,y=corr_df['Sales'][i]+0.3,s=corr_df['Employee'][i], 
          fontdict=dict(color='red',size=10),
          bbox=dict(facecolor='yellow',alpha=0.5))
```

```python
corr_df
```

```python
corr_matrix = np.corrcoef(corr_df['Number of Territories'], corr_df.Sales)
corr = corr_matrix[0,1]
R_sq = corr**2
 
print(R_sq)
```
